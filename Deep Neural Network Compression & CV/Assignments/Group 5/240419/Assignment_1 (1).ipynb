````python

"""
swap_mag_phase.py

Swap magnitude and phase between two images, reconstruct and save/display results.

Requirements:
    pip install numpy pillow matplotlib

Usage:
    - Edit img1_path and img2_path to point to your images (they may be grayscale or color).
    - Run: python swap_mag_phase.py
"""

from pathlib import Path
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt


img1_path = Path("/content/a60d3db3-68e3-4d67-8431-ef36b06da778.jpg")
img2_path = Path("/content/pexels-ekrulila-31671970.jpg")



def load_gray_float(path: Path, resize_to=None):
    img = Image.open(path).convert("L")
    if resize_to:
        img = img.resize(resize_to, Image.LANCZOS)
    arr = np.array(img).astype(np.float64)
    return arr, img


def reconstruct_from_mag_phase(mag_shifted, phase_shifted):

    complex_shifted = mag_shifted * np.exp(1j * phase_shifted)

    complex_unshifted = np.fft.ifftshift(complex_shifted)

    recon = np.fft.ifft2(complex_unshifted)
    recon_real = np.real(recon)

    recon_norm = recon_real - recon_real.min()
    if recon_norm.max() > 0:
        recon_norm = recon_norm / recon_norm.max() * 255.0
    return recon_norm.astype(np.uint8)

def main():

    arr1, img1_pil = load_gray_float(img1_path)


    h1, w1 = arr1.shape

    arr2, img2_pil = load_gray_float(img2_path, resize_to=(w1, h1))


    F1 = np.fft.fftshift(np.fft.fft2(arr1))
    F2 = np.fft.fftshift(np.fft.fft2(arr2))


    mag1 = np.abs(F1)
    phase1 = np.angle(F1)

    mag2 = np.abs(F2)
    phase2 = np.angle(F2)


    recon1 = reconstruct_from_mag_phase(mag1, phase2)
    recon2 = reconstruct_from_mag_phase(mag2, phase1)
    out_dir = Path("outputs")
    out_dir.mkdir(exist_ok=True)
    out1_path = out_dir / "recon_mag1_phase2.png"
    out2_path = out_dir / "recon_mag2_phase1.png"

    Image.fromarray(recon1).save(out1_path)
    Image.fromarray(recon2).save(out2_path)
    print(f"Saved: {out1_path}")
    print(f"Saved: {out2_path}")


    fig, axes = plt.subplots(2, 3, figsize=(14, 8))

    axes[0,0].imshow(img1_pil, cmap="gray")
    axes[0,0].set_title("Original Image 1")
    axes[0,0].axis("off")

    axes[0,1].imshow(img2_pil, cmap="gray")
    axes[0,1].set_title("Original Image 2")
    axes[0,1].axis("off")


    axes[0,2].imshow(np.log(mag1 + 1), cmap="gray")
    axes[0,2].set_title("Log |F1|")
    axes[0,2].axis("off")

    axes[1,0].imshow(recon1, cmap="gray")
    axes[1,0].set_title("Magnitude1 + Phase2")
    axes[1,0].axis("off")

    axes[1,1].imshow(recon2, cmap="gray")
    axes[1,1].set_title("Magnitude2 + Phase1")
    axes[1,1].axis("off")

    axes[1,2].imshow(np.log(mag2 + 1), cmap="gray")
    axes[1,2].set_title("Log |F2|")
    axes[1,2].axis("off")

    plt.tight_layout()
    plt.show()



if __name__ == "__main__":
    main()

````
