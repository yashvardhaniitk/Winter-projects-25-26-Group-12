NOW ONTO THE HISTOGRAM CREATION PART


````python
def grayscale_histogram(img):
    #  initialize an array of 256 zeros to store counts for each intensity level
    hist = np.zeros(256, dtype=int)
    # iterate through every single pixel in the flattened image array
    for pixel in img.ravel():
        hist[pixel] += 1
    return hist

hist_1 = grayscale_histogram(img_gray)

plt.figure(figsize=(8,4))
plt.plot(hist_1)
plt.xlabel("Intensity (0â€“255)")
plt.ylabel("Frequency")
plt.title("Grayscale Histogram")
plt.show()
````

````python
def rgb_histogram(img) : 
    hist_r = grayscale_histogram(img[:,:,0])
    hist_g = grayscale_histogram(img[:,:,1])
    hist_b = grayscale_histogram(img[:,:,2])
    return hist_r, hist_g, hist_b     
````

````python
hist_r, hist_g, hist_b = rgb_histogram(img_data)

plt.plot(hist_r, color = 'Red')
plt.plot(hist_g, color = "Green")
plt.plot(hist_b, color = "Blue")
````

Write a function <code>rgb_to_hsv</code> that takes a 0-255 ( not normalised ) RGB image and returns an HSV image ( also not normalised ).

````python
import numpy as np

def rgb_2_hsv(img):
    R = img[:, :, 0].astype(float)
    G = img[:, :, 1].astype(float)
    B = img[:, :, 2].astype(float)

    cmax = np.maximum(np.maximum(R, G), B)
    cmin = np.minimum(np.minimum(R, G), B)
    delta = cmax - cmin

    ## HUE
    H = np.zeros_like(cmax, dtype=float)

    # avoid division by zero
    delta_safe = np.where(delta == 0, 1.0, delta)

    h_r = 60.0 * np.mod((G - B) / delta_safe, 6.0)
    h_g = 60.0 * (((B - R) / delta_safe) + 2.0)
    h_b = 60.0 * (((R - G) / delta_safe) + 4.0)

    H = np.where(delta == 0, 0.0, H)
    H = np.where((delta != 0) & (cmax == R), h_r, H)
    H = np.where((delta != 0) & (cmax == G), h_g, H)
    H = np.where((delta != 0) & (cmax == B), h_b, H)

    # ensure H in [0, 360)
    H = np.mod(H, 360.0)

    ## SATURATION 
    S = np.where(cmax == 0, 0.0, (delta / cmax) * 255.0)
    # clip to numeric range
    S = np.clip(S, 0.0, 255.0)

    ## VALUE 
    V = cmax.copy()
    V = np.clip(V, 0.0, 255.0)

    # sanitize any non-finite values just in case
    H[~np.isfinite(H)] = 0.0
    S[~np.isfinite(S)] = 0.0
    V[~np.isfinite(V)] = 0.0

    return np.dstack((H, S, V))
````

````python
img_hsv_data = rgb_2_hsv(img_data)
````

````python
img_hsv_data.shape
````

### MEXICOFY FILTER ðŸ¤ 

````python
def mexico_fy(img_rgb):
    # Ensure image is float for calculations
    img = img_rgb / 255.0

    # Increasing yellowishness:
    # Increase red and green channels more, decrease blue slightly
    img_tinted = img.copy()
    img_tinted[:, :, 0] = np.clip(img[:, :, 0] * 1.3, 0, 1) # Boost Red
    img_tinted[:, :, 1] = np.clip(img[:, :, 1] * 1.2, 0, 1) # Boost Green
    img_tinted[:, :, 2] = np.clip(img[:, :, 2] * 0.7, 0, 1) # Reduce Blue

    # Increasing contrast
    contrast_factor = 1.3
    img_contrast = 0.5 + contrast_factor * (img_tinted - 0.5)
    img_contrast = np.clip(img_contrast, 0, 1)

    # Saturation Boost (using HSV conversion via cv2):
    # Convert to HSV, boost saturation, convert back to RGB
    img_hsv = cv2.cvtColor((img_contrast * 255).astype(np.uint8), cv2.COLOR_RGB2HSV)
    saturation_factor = 1.2
    img_hsv[:, :, 1] = np.clip(img_hsv[:, :, 1] * saturation_factor, 0, 255) # S channel
    filtered_img_rgb = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB)

    return filtered_img_rgb.astype(np.uint8)
````

### Harry Potter & The Prisoner Of Azkaban Filter ðŸ§™ðŸª„

````python
def POA_filter(img_rgb):
    img = img_rgb.astype(np.float32) / 255.0 # Normalizing

    # Brightness Decrease
    brightness_offset = -0.25 # Subtract to darken
    img_dark = np.clip(img + brightness_offset, 0, 1)

    # Contrast Increase
    contrast_factor = 1.3
    img_contrast = 0.5 + contrast_factor * (img_dark - 0.5)
    img_contrast = np.clip(img_contrast, 0, 1)

    # Cool Color Tint
    # Increase blue, slightly decrease red and green
    img_tinted = img_contrast.copy()
    img_tinted[:, :, 0] = np.clip(img_contrast[:, :, 0] * 0.9, 0, 1) # Reduce Red
    img_tinted[:, :, 1] = np.clip(img_contrast[:, :, 1] * 0.95, 0, 1) # Reduce Green
    img_tinted[:, :, 2] = np.clip(img_contrast[:, :, 2] * 1.5, 0, 1) # Boost Blue

    # 4. Desaturation (using HSV conversion via cv2):
    img_hsv = cv2.cvtColor((img_tinted * 255).astype(np.uint8), cv2.COLOR_RGB2HSV)
    saturation_factor = 0.7 # Reduce saturation
    img_hsv[:, :, 1] = np.clip(img_hsv[:, :, 1] * saturation_factor, 0, 255) # S channel
    filtered_img_rgb = cv2.cvtColor(img_hsv, cv2.COLOR_HSV2RGB)

    return filtered_img_rgb.astype(np.uint8)
````

#### White Balance Function
Write a function <code>white_patch_balance(img_rgb)</code> that takes a 0-255 RGB image and then returns a White Patch Method Balanced 0-255 RBG image. Use Numpy only to keep track of the brightest pixel and then scale the entire image with the factor you find. Display the original and output image.

````python
# Your PART D codes follow here. You can create more cells under this. Write clean and commented code.
````

````python
def white_patch_balance(img):

    # Tracking the maximum R-G-B channel values throughout the image
    max_r = np.max(img[:, :, 0])
    max_g = np.max(img[:, :, 1])
    max_b = np.max(img[:, :, 2])


    # We want the max value in each channel to become 255
    scale_r = 255.0 / max_r
    scale_g = 255.0 / max_g
    scale_b = 255.0 / max_b

    # Apply scaling to each channel
    balanced_img = img.copy()
    balanced_img[:, :, 0] = balanced_img[:, :, 0] * scale_r
    balanced_img[:, :, 1] = balanced_img[:, :, 1] * scale_g
    balanced_img[:, :, 2] = balanced_img[:, :, 2] * scale_b

    return balanced_img
````

````python
wpb = white_patch_balance(img_data)
plt.imshow(wpb)
plt.axis("off")
````
