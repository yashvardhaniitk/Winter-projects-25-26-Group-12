```python
import sys

sys.setrecursionlimit(200005)
def solveq1():
    input = sys.stdin.read
    data = input().split()
    iterator = iter(data)

    try:
        n = int(next(iterator))
        m = int(next(iterator))
    except StopIteration:
        return

    adj = [[] for _ in range(n + 1)]
    for _ in range(m):
        u = int(next(iterator))
        v = int(next(iterator))
        adj[u].append(v)
        adj[v].append(u)

    visited = [False] * (n + 1)

    def dfs(u, p):
        visited[u] = True
        for v in adj[u]:
            if v == p:
                continue
            if visited[v]:
                return True
            if dfs(v, u):
                return True
        return False

    hasCycle = False

    for i in range(1, n + 1):
        if not visited[i]:
            if dfs(i, -1):
                hasCycle = True
                break

    if hasCycle:
        print("YES")
    else:
        print("NO")
    solveq1()
```

```python
import sys
from collections import deque

sys.setrecursionlimit(200005)

def solveq2():
    input = sys.stdin.read
    data = input().split()
    iterator = iter(data)

    try:
        n = int(next(iterator))
        m = int(next(iterator))
    except StopIteration:
        return

    adj = [[] for _ in range(n + 1)]
    for _ in range(m):
        u = int(next(iterator))
        v = int(next(iterator))
        adj[u].append(v)
        adj[v].append(u)

    color = [-1] * (n + 1)
    isBipartite = True

    for i in range(1, n + 1):
        if color[i] == -1:
            q = deque([i])
            color[i] = 0

            while q:
                u = q.popleft()

                for v in adj[u]:
                    if color[v] == -1:
                        color[v] = 1 - color[u]
                        q.append(v)
                    elif color[v] == color[u]:
                        isBipartite = False
                        break
                if not isBipartite:
                    break
        if not isBipartite:
            break

    if isBipartite:
        print("YES")
        print(*(color[1:]))
    else:
        print("NO")

    solveq2()
```

```python
import sys
import heapq

def solveq3():
    input = sys.stdin.read
    data = input().split()
    iterator = iter(data)

    try:
        n = int(next(iterator))
        m = int(next(iterator))
    except StopIteration:
        return

    adj = [[] for _ in range(n + 1)]
    for _ in range(m):
        u = int(next(iterator))
        v = int(next(iterator))
        w = int(next(iterator))
        adj[u].append((v, w))

    dist = [float('inf')] * (n + 1)
    dist[1] = 0
    pq = [(0, 1)]

    while pq:
        d, u = heapq.heappop(pq)

        if d > dist[u]:
            continue

        for v, w in adj[u]:
            if dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
                heapq.heappush(pq, (dist[v], v))

    if dist[n] == float('inf'):
        print("-1")
    else:
        print(dist[n])

    solveq3()
```

```python
import sys
import heapq

def dijkstra(start_node, graph, n):
    dist = [float('inf')] * (n + 1)
    dist[start_node] = 0
    pq = [(0, start_node)]

    while pq:
        d, u = heapq.heappop(pq)
        if d > dist[u]:
            continue
        for v, w in graph[u]:
            if dist[u] + w < dist[v]:
                dist[v] = dist[u] + w
                heapq.heappush(pq, (dist[v], v))
    return dist

def solveq4():
    input = sys.stdin.read
    data = input().split()
    iterator = iter(data)

    try:
        n = int(next(iterator))
        m = int(next(iterator))
        k = int(next(iterator))
    except StopIteration:
        return

    adj = [[] for _ in range(n + 1)]
    rev_adj = [[] for _ in range(n + 1)]

    for _ in range(m):
        u = int(next(iterator))
        v = int(next(iterator))
        w = int(next(iterator))
        adj[u].append((v, w))
        rev_adj[v].append((u, w))

    waypoints = []
    for _ in range(k):
        waypoints.append(int(next(iterator)))

    dist1 = dijkstra(1, adj, n)
    distn = dijkstra(n, rev_adj, n)

    minEntry = float('inf')
    for wp in waypoints:
        if dist1[wp] < minEntry:
            minEntry = dist1[wp]

    min_exit = float('inf')
    for wp in waypoints:
        if distn[wp] < min_exit:
            min_exit = distn[wp]

    ans = dist1[n]

    if minEntry != float('inf') and min_exit != float('inf'):
        hyperloopCost = minEntry + min_exit
        if hyperloopCost < ans:
            ans = hyperloopCost

    if ans == float('inf'):
        print("-1")
    else:
        print(ans)

    solveq4()
```
