```python
import numpy as np
import matplotlib.pyplot as plt

def gray_histogram(gray_img):
    """
    gray_img : 2D array, values 0255 (uint8 preferred)
    Returns: 1D array hist of length 256
    """
    hist = np.zeros(256, dtype=np.int32)


    rows, cols = gray_img.shape
    for i in range(rows):
        for j in range(cols):
            pixel = gray_img[i, j]     # 0255
            hist[pixel] += 1

    return hist

h = gray_histogram(gray_numpy)

plt.figure(figsize=(6,4))
plt.plot(h, color='black')
plt.title("Grayscale Histogram")
plt.xlabel("Intensity Value (0255)")
plt.ylabel("Count")
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

```python
def rgb_histogram(rgb):
    """
    rgb_img : H  W  3 image, values 0255
    Returns:
        histR, histG, histB  (each length 256)
    """
    histR = np.zeros(256, dtype=np.int32)
    histG = np.zeros(256, dtype=np.int32)
    histB = np.zeros(256, dtype=np.int32)

    rows, cols, _ = rgb.shape

    for i in range(rows):
        for j in range(cols):
            r, g, b = rgb[i, j]
            histR[r] += 1
            histG[g] += 1
            histB[b] += 1

    return histR, histG, histB
histR, histG, histB = rgb_histogram(rgb)

plt.figure(figsize=(8,5))
plt.plot(histR, color='red', label='Red Channel')
plt.plot(histG, color='green', label='Green Channel')
plt.plot(histB, color='blue', label='Blue Channel')

plt.title("RGB Histogram ")
plt.xlabel("Intensity Value (0255)")
plt.ylabel("Count")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.4)
plt.show()
```

```python
import numpy as np

def rgb_to_hsv(rgb):
    """
    Input:
        rgb image in uint8, shape (H, W, 3), values 0255
    Output:
        hsv image in uint8, shape (H, W, 3)
        H  [0,180], S  [0,255], V  [0,255]
        Compatible with OpenCV cvtColor
    """

    rgb = rgb.astype(np.float32)
    R = rgb[...,0] / 255.0
    G = rgb[...,1] / 255.0
    B = rgb[...,2] / 255.0

    # Value
    V = np.max(rgb/255.0, axis=-1)

    # Delta
    Cmax = np.max(rgb/255.0, axis=-1)
    Cmin = np.min(rgb/255.0, axis=-1)
    delta = Cmax - Cmin

    # Hue calculation
    H = np.zeros_like(Cmax)

    # Avoid division by zero
    mask = delta != 0

    # If Cmax = R
    idx = (Cmax == R) & mask
    H[idx] = (((G[idx] - B[idx]) / delta[idx]) % 6)

    # If Cmax = G
    idx = (Cmax == G) & mask
    H[idx] = ((B[idx] - R[idx]) / delta[idx]) + 2

    # If Cmax = B
    idx = (Cmax == B) & mask
    H[idx] = ((R[idx] - G[idx]) / delta[idx]) + 4

    H = H * 60  # degrees, 0360

    # Saturation
    S = np.zeros_like(Cmax)
    S[mask] = delta[mask] / Cmax[mask]

    # --- OpenCV Scaling ---
    # H : 0360   0180 (stored as uint8)
    # S : 01     0255
    # V : 01     0255
    H = (H / 2).astype(np.uint8)
    S = (S * 255).astype(np.uint8)
    V = (V * 255).astype(np.uint8)

    hsv = np.stack([H, S, V], axis=-1)
    return hsv

import cv2
import matplotlib.pyplot as plt

# Load BGR  convert to RGB first
rgb = cv2.cvtColor(cv2.imread("dog.jpg"), cv2.COLOR_BGR2RGB)

# Convert using OUR function
hsv_custom = rgb_to_hsv(rgb)

# Convert back to RGB using OpenCV
rgb_reconstructed = cv2.cvtColor(hsv_custom, cv2.COLOR_HSV2RGB)

plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.imshow(rgb)
plt.title("Original RGB")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(rgb_reconstructed)
plt.title("Reconstructed RGB ")
plt.axis("off")

plt.show()

```

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# ---------------------------
# Coffee Filter (old manuscript)
# ---------------------------
def coffee_filter(img_rgb, strength=0.9):
    """
    Coffee Filter:
    - img_rgb: HxWx3 uint8 RGB (0..255)
    - strength: [0..1] how strong the effect (1.0 full)
    Returns: HxWx3 uint8 RGB
    """
    img = img_rgb.astype(np.float32)

    # 1) Slight desaturate (make it look aged)
    hsv = cv2.cvtColor(np.clip(img,0,255).astype(np.uint8), cv2.COLOR_RGB2HSV).astype(np.float32)
    hsv[...,1] *= (1.0 - 0.25*strength)           # reduce saturation
    hsv[...,1] = np.clip(hsv[...,1], 0, 255)
    img1 = cv2.cvtColor(hsv.astype(np.uint8), cv2.COLOR_HSV2RGB).astype(np.float32)

    # 2) Warm/brown tint (channel mixing)
    #    increase R a bit, G moderate, reduce Blue -> brownish
    tint = np.array([0.88 + 0.24*strength,  # B channel scale (reduce blue)
                     0.95 + 0.12*strength,  # G channel scale
                     1.02 + 0.30*strength]) # R channel scale (boost red)
    img2 = img1 * tint[np.newaxis, np.newaxis, :]

    # 3) Add paper overlay (soft multiply/blend with brown color)
    brown_layer = np.full_like(img2, [180, 150, 95], dtype=np.float32)  # RGB brownish
    alpha = 0.12 * strength
    img3 = img2 * (1 - alpha) + brown_layer * alpha

    # 4) Vignette (darken corners slightly)
    h, w = img3.shape[:2]
    # create radial mask
    Y, X = np.ogrid[:h, :w]
    cy, cx = h/2, w/2
    dist = np.sqrt(((Y - cy)/ (h/2))**2 + ((X - cx)/(w/2))**2)
    vignette = 1.0 - 0.6 * (dist**2) * strength    # center ~1, corners down
    vignette = np.clip(vignette, 0.4, 1.0)[..., np.newaxis]
    img4 = img3 * vignette

    # 5) Add small film grain
    noise_strength = 6.0 * strength
    noise = (np.random.randn(h, w, 1) * noise_strength).astype(np.float32)
    img5 = img4 + noise

    # 6) Final tone/gamma (slightly lower contrast, gentle curve)
    img5 = np.clip(img5, 0, 255)
    gamma = 1.02 - 0.12*strength
    img_final = 255.0 * ((img5 / 255.0) ** (1.0 / gamma))

    return np.clip(img_final, 0, 255).astype(np.uint8)


# ---------------------------
# Swan Lake (pearly, cold)
# ---------------------------
def swan_lake(img_rgb, strength=0.9):
    """
    Swan Lake:
    - pearly, cold, soft-glow look
    - strength [0..1]
    """
    img = img_rgb.astype(np.float32)

    # 1) Slight desaturate (pearly tends to be less saturated)
    hsv = cv2.cvtColor(np.clip(img,0,255).astype(np.uint8), cv2.COLOR_RGB2HSV).astype(np.float32)
    hsv[...,1] *= (1.0 - 0.20*strength)
    hsv[...,1] = np.clip(hsv[...,1], 0, 255)
    img1 = cv2.cvtColor(hsv.astype(np.uint8), cv2.COLOR_HSV2RGB).astype(np.float32)

    # 2) Cool / pearly tint: channel mixing: boost blue/cyan, subtle magenta-ish in highlights
    #    Use per-channel scales and a highlight mask
    scale = np.array([0.96 - 0.04*strength,   # B channel in array ordering is index 0 (we use RGB arrays)
                      1.00 + 0.04*strength,   # G
                      1.04 + 0.10*strength])  # R (we'll adjust with highlight map later)
    # but we want cooler overall -> boost blue and cyan; do channel mix differently:
    # We'll compute a linear combination to shift towards cyan/blue
    blue_boost = np.array([1.06 + 0.12*strength, 1.03 + 0.06*strength, 0.97 - 0.02*strength])
    img2 = img1 * blue_boost[np.newaxis, np.newaxis, :]

    # 3) Create a soft glow: blur bright areas and screen-blend
    bright = np.clip(img2 - 200, 0, 55)    # highlight extractor (values near white)
    # upscale bright to emphasize
    bright_up = bright * (2.0 * strength)
    # blur the bright map to create glow
    glow = cv2.GaussianBlur(bright_up.astype(np.uint8), (0,0), sigmaX=15*strength, sigmaY=15*strength).astype(np.float32)
    # screen blend: result = 255 - (255 - a)*(255 - b)/255
    a = img2
    b = glow
    img3 = 255.0 - ((255.0 - a) * (255.0 - b) / 255.0)

    # 4) Lift shadows slightly and reduce contrast (pearly soft look)
    #    contrast factor <1 reduces contrast
    contrast = 1.0 - 0.12*strength
    mid = 128.0
    img4 = (img3 - mid) * contrast + mid
    # lift shadows
    shadow_lift = 12.0 * strength
    img4 = img4 + shadow_lift

    # 5) Slight bluish-white color grade: mix a tiny amount of pale cyan
    pale_cyan = np.full_like(img4, [220, 235, 245], dtype=np.float32)  # RGB pale cyan/pearl
    blend_alpha = 0.08 * strength
    img5 = img4 * (1 - blend_alpha) + pale_cyan * blend_alpha

    # 6) Subtle vignette lightening center (not darkening)
    h, w = img5.shape[:2]
    Y, X = np.ogrid[:h, :w]
    cy, cx = h/2, w/2
    dist = np.sqrt(((Y - cy)/(h/2))**2 + ((X - cx)/(w/2))**2)
    vign = 1.0 + 0.25 * (1 - np.clip(dist, 0, 1)) * strength   # center brightening
    img6 = img5 * vign[..., np.newaxis]

    # 7) Final gentle gamma (soften)
    img6 = np.clip(img6, 0, 255)
    gamma = 1.0 + 0.08*strength
    img_final = 255.0 * ((img6 / 255.0) ** (1.0 / gamma))

    return np.clip(img_final, 0, 255).astype(np.uint8)


# ---------------------------
# Example usage & display
# ---------------------------
# Replace the following line with your image loading. It expects an RGB image.
# Example:
# img = cv2.cvtColor(cv2.imread('your_image.jpg'), cv2.COLOR_BGR2RGB)

# For demo purpose, try to use variable 'img' if present; otherwise create a small placeholder:
try:
    img  # if the user already has img in the session
except NameError:
    # create a warm demo if not provided
    h, w = 480, 720
    # simple scenic-like gradient for demo
    R = np.tile(np.linspace(200, 80, w), (h,1)).astype(np.uint8)
    G = np.tile(np.linspace(190, 60, w), (h,1)).astype(np.uint8)
    B = np.tile(np.linspace(170, 40, w), (h,1)).astype(np.uint8)
    img = np.dstack([R, G, B])

# Run filters
coffee = coffee_filter(rgb, strength=0.95)
swan = swan_lake(rgb, strength=0.95)

# Show original and filtered images side-by-side
plt.figure(figsize=(15,6))
plt.subplot(1,3,1); plt.imshow(img); plt.title("Original"); plt.axis('off')
plt.subplot(1,3,2); plt.imshow(coffee); plt.title("Coffee Filter (manuscript)"); plt.axis('off')
plt.subplot(1,3,3); plt.imshow(swan); plt.title("Swan Lake (pearly cold)"); plt.axis('off')
plt.tight_layout()
plt.show()

# ---------------------------
# Short notes (2-3 lines each)
# ---------------------------
print("Coffee Filter: adds warm brown tint, desaturates slightly, adds vignette and grain  good for portraits, vintage scenes, old manuscript or coffee-stain aesthetics.")
print("Swan Lake: adds pearly cool tint, soft glow, lifted shadows and subtle desaturation  good for portraits, winter/ice scenes, fashion/editorial, dreamy landscapes.")
```
